# Project description
- To build an end to end image classifier (simple object detection)
- It is a binary classifier with classes: **car** and **bike** (given an image tells if it is a car or a bike)
- Data is hand collected, cleaned and prepared (in data folder of this project)
- Took related pictures manually with a camera but a few images are downloaded from the internet to increase the diversity of the data
- Following important steps are done
    1. Data collection: **raw unlabelled data** is collected (images folder)
    2. Data cleaning: made the data suitable for this project
    3. **Clustering**: segregated images using **KMeans clustering** (in folders cluster0 and cluster1) and **labelled data manually** (in folders(class0 and class1)
    4. Feature Engineering (for different classifiers)
    5. Built different classifiers for the data prepared
    6. Three classifiers (**SVM, Random Forest amd Neural Network**) are built for the same data
    7. **Grid Search Cross Validation** used to do **hyperparameter search** for all the three models
    8. Classification metrics are calculated (**accuracy, confusion matrix, PR curve and ROC curve, AUC**)
    9. Models are compared based on the classification metrics
    10. All models are saved (models folder of this project) and are used for prediction
    11. A **webapp** using these models is built and hosted at https://image-classifier-tool.herokuapp.com/
    12. Drag and drop a image from the internet or upload a relevant image from your local system to test
- **Note for testing**: Almost all the bike images used in this project have the following characteristics: they are in driving mode mounted by one or more people and they are captured from the side view. Resting, front view and back view images are very minimal so, while testing for images they should be modified to meet these characteristics to yield proper results

# Setup

```python
# setup paths

DATADIR = "./data/" # data directory
MODELDIR = "./models/" # models directory

IMAGEDIR = DATADIR + "images/" # images directory
CLUSTER0DIR = DATADIR + "cluster0/"
CLUSTER1DIR = DATADIR + "cluster1/"
CLASS0DIR = DATADIR + "class0/"
CLASS1DIR = DATADIR + "class1/"

IMGSIZE = 128
```

```python
# import section

import os
import shutil
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.transform import resize
from seaborn_image import ImageGrid

from sklearn.cluster import KMeans
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import silhouette_score

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

import pickle

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_recall_curve, roc_curve, auc

import warnings
warnings.filterwarnings("ignore")
```

# Data Collection and Preparation

Our data contains images of 2 types (so this project deals with binary image classification/clustering activities)

```python
# raw unlabelled data - images in IMAGEDIR

# all images from both the classes are in a single folder
# Images of both the classes are mixed and unlabelled
# so the first step is to segregate them into different folders according to their classes
# for this we perform clustering and segregate images of different clusters into respective folders
```

## Load Raw Data

```python
images = []
paths = []

files = os.listdir(IMAGEDIR)
for num, image_name in enumerate(files):
    path = IMAGEDIR + image_name
    try:
        image = resize(imread(path, as_gray=True), (IMGSIZE, IMGSIZE))
        images.append(image)
        paths.append(path)
    except:
        print(image_name)
```

```python
imshow(images[10], origin="upper")
```

## Clustering

```python
# Perform clustering before manual labelling
# get features for clustering algorithm (KMeans)

features = []

for image in images:
    features.append(image.flatten())
```

```python
# find the best number of clusters (k)

sil = []
kl = []

k_max = 10

for k in range(2, k_max+1):
    kmeans = KMeans(n_clusters = k).fit(features)
    labels = kmeans.labels_
    sil.append(silhouette_score(features, labels, metric="euclidean"))
    kl.append(k)

plt.plot(kl, sil)
plt.ylabel("Silhoutte Score")
plt.xlabel("K")
plt.show()
```

```python
# from the above figure k = 2 is best
# performing clustering with kmeans, k = 2

k = 2
kmeans = KMeans(n_clusters=k)
kmeans.fit(features)
predictions = kmeans.predict(features)

for i in range(k):
    PATH = DATADIR + "cluster" + str(i)
    if os.path.exists(PATH):
        shutil.rmtree(PATH)
    os.mkdir(PATH)
    
# segregating images in different clusters into respective folders cluster0/ and cluster1/

for i in range(len(paths)):
    shutil.copy2(paths[i], DATADIR + "cluster" +str(predictions[i]))
```

```python
# creating copies of the cluster folders named class0/ and class1/
# images in this folder are manullay examined and removed to retain onlly the images that actually belong to that class
# example: if class0 has large number of bike images then all the car images are removed from that folder

if os.path.exists(CLASS0DIR):
    shutil.rmtree(CLASS0DIR)

if os.path.exists(CLASS1DIR):
    shutil.rmtree(CLASS1DIR)

shutil.copytree(CLUSTER0DIR, CLASS0DIR)
shutil.copytree(CLUSTER1DIR, CLASS1DIR)

print("finished creating classes")
```

# Data Preprocessing

```python
# load data from both the folders class0 and class1

images = []
labels = []
image_names0 = os.listdir(CLASS0DIR)
image_names1 = os.listdir(CLASS1DIR)

for image_name in image_names0:
    path = IMAGEDIR + image_name
    image = resize(imread(path, as_gray=True), (IMGSIZE, IMGSIZE))
    images.append(image)
    labels.append(0)
    
for image_name in image_names1:
    path = IMAGEDIR + image_name
    image = resize(imread(path, as_gray=True), (IMGSIZE, IMGSIZE))
    images.append(image)
    labels.append(1)
    
img_dict = {
    "image": images,
    "label": labels
}

img_df = pd.DataFrame(img_dict)
```

# Building Models

## Model 1 - SVM

```python
# load data

file = DATADIR + "images.pkl"

with open(file, "rb") as fp:
    img_df = pickle.load(fp)
    
images = img_df["image"].values.tolist()
labels = img_df["label"].values.tolist()

# feature engineering

svm_features = []

for image in images:
    svm_features.append(image.flatten())

# preparing training and test sets

x_train, x_test, y_train, y_test = train_test_split(svm_features, labels, test_size=0.3, random_state=50)

# model

svc = svm.SVC(probability=True)

# training along with hyperparameter search

param_grid = [
              {"C":[1, 10, 100], "kernel":["linear"]},
              {"C":[1, 10, 100], "gamma":[0.001, 0.0001], "kernel":["rbf"]}
]

clf = GridSearchCV(svc, param_grid)
clf.fit(x_train, y_train)

# predictions for training set

y_pred_train = clf.predict(x_train)

# predictions for test set

y_pred_test = clf.predict(x_test)

# training metrics

print("Training metrics:")
print(classification_report(y_true= y_train, y_pred= y_pred_train))

# test metrics

print("Test metrics:")
print(classification_report(y_true= y_test, y_pred= y_pred_test))
```

### Accuracy and Confusion Matrix

```python
print("accuracy:", accuracy_score(y_pred_test, y_test))
print("confusion matrix:\n", confusion_matrix(y_pred_test, y_test))
```

### PR Curve

```python
# pr curve

y_score = clf.predict_proba(x_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_score)
plt.plot(recall,precision,"*")
plt.plot(recall,precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("PR Curve")
```

### ROC Curve

```python
# roc curve

fpr, tpr, thresholds = roc_curve(y_true=y_test,y_score=y_score)
auc_score = auc(fpr,tpr)
plt.title("ROC Curve" + "(AUC=" + str(auc_score) + ")")
plt.plot(fpr, tpr,'*')
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

### Save and use the model for prediction

```python
# save model

file = MODELDIR + "svm.pkl"

with open(file, "wb") as fp:
    pickle.dump(clf, fp)
```

```python
# use the saved model to predict
# note image url address should be given and the address usually ends with .jpg, .png etc., or the size of the image file

svm = 0

file = MODELDIR + "svm.pkl"

with open(file, "rb") as fp:
    svm = pickle.load(fp)
    
classes = ["bike", "car"]
test_features = []
url = input("Enter your URL ")
try:
    img = imread(url, as_gray=True)
    img_resized = resize(img, (IMGSIZE, IMGSIZE))
    test_features.append(img_resized.flatten())

    imshow(img_resized, origin="upper")
    y_out = svm.predict(test_features)
    prob = svm.predict_proba(test_features)
    print(f"Predicted output: {classes[int(y_out)]} with {prob[0][int(y_out)]*100:0.4f}% confidence")
except Exception as e:
    print(e)
```

## Model 2 - Random Forest

```python
# load data

file = DATADIR + "images.pkl"

with open(file, "rb") as fp:
    img_df = pickle.load(fp)
    
images = img_df["image"].values.tolist()
labels = img_df["label"].values.tolist()

# feature engineering

rf_features = []

for image in images:
    rf_features.append(image.flatten())

# preparing training and test sets

x_train, x_test, y_train, y_test = train_test_split(rf_features, labels, test_size=0.3, random_state=50)

# model

rf = RandomForestClassifier()

# training along with hyperparameter search

param_grid = {
    "n_estimators": [30,60],
    "min_samples_leaf": [2,3]
}

clf = GridSearchCV(rf, param_grid)
clf.fit(x_train, y_train)

# predictions for training set

y_pred_train = clf.predict(x_train)

# predictions for test set

y_pred_test = clf.predict(x_test)

# training metrics

print("Training metrics:")
print(classification_report(y_true= y_train, y_pred= y_pred_train))

# test metrics

print("Test metrics:")
print(classification_report(y_true= y_test, y_pred= y_pred_test))
```

### Accuracy and Confusion Matrix

```python
print("accuracy:", accuracy_score(y_pred_test, y_test))
print("confusion matrix:\n", confusion_matrix(y_pred_test, y_test))
```

### PR Curve

```python
# pr curve

y_score = clf.predict_proba(x_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_score)
plt.plot(recall,precision,"*")
plt.plot(recall,precision)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.xlim([-0.1,1.05])
plt.ylim([-0.1,1.05])
plt.title("PR Curve")
```

### ROC Curve

```python
# roc curve

fpr, tpr, thresholds = roc_curve(y_true=y_test,y_score=y_score)
auc_score = auc(fpr,tpr)
plt.title("ROC Curve" + "(AUC=" + str(auc_score) + ")")
plt.plot(fpr, tpr,'*')
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

### Save and use the model for prediction

```python
# save model

file = MODELDIR + "rf.pkl"

with open(file, "wb") as fp:
    pickle.dump(clf, fp)
```

## Model 3 - Neural Network

```python
# load data

file = DATADIR + "images.pkl"

with open(file, "rb") as fp:
    img_df = pickle.load(fp)
    
images = img_df["image"].values.tolist()
labels = img_df["label"].values.tolist()

# feature engineering

mlp_features = []

for image in images:
    mlp_features.append(image.flatten())
```

```python
# preparing training and test sets

x_train, x_test, y_train, y_test = train_test_split(mlp_features, labels, test_size=0.2, random_state=50)
```

```python
# define model

mlp = MLPClassifier(max_iter=100)
```

```python
# training along with hyperparameter search

param_grid = {
    "hidden_layer_sizes": [(10,10,10), (10,20,10), (20,)],
    "activation": ['tanh', 'relu'],
    "solver": ['sgd', 'adam'],
    "alpha": [0.0001, 0.05],
    "learning_rate": ['constant','adaptive'],
}

clf = GridSearchCV(mlp, param_grid, n_jobs=-1, cv=3)
clf.fit(x_train, y_train)
```

```python
# predictions for training set

y_pred_train = clf.predict(np.array(x_train, dtype=int))

# predictions for test set

y_pred_test = clf.predict(np.array(x_test, dtype=int))

# training metrics

print("Training metrics:")
print(classification_report(y_true= y_train, y_pred= y_pred_train))

# test metrics

print("Test metrics:")
print(classification_report(y_true= y_test, y_pred= y_pred_test))
```

### Accuracy and Confusion Matrix

```python
print("accuracy:", accuracy_score(y_pred_test, y_test))
print("confusion matrix:\n", confusion_matrix(y_pred_test, y_test))
```

### PR Curve

```python
# pr curve

y_score = clf.predict_proba(x_test)[:,1]
precision, recall, thresholds = precision_recall_curve(y_test, y_score)
plt.plot(recall,precision,"*")
plt.plot(recall,precision)

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.xlim([-0.1,1.05])
plt.ylim([-0.1,1.05])
plt.title("PR Curve")
```

### ROC Curve

```python
# roc curve

fpr, tpr, thresholds = roc_curve(y_true=y_test,y_score=y_score)
auc_score = auc(fpr,tpr)
plt.title("ROC Curve" + "(AUC=" + str(auc_score) + ")")
plt.plot(fpr, tpr,'*')
plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```

### Save and use the model for prediction

```python
# save model

file = MODELDIR + "nn.pkl"

with open(file, "wb") as fp:
    pickle.dump(clf, fp)
```
